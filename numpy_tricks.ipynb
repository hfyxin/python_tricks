## plot formatting
Figure size, subplots, legends, ...

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([0, 1,  2,  3, 3.5, 4,  7, 9.5, 10])
y1 = np.array([0, 1.5,8, -3, 0.2, 1.3,6,6.18, 1])
y2 = y1+1

plt.figure(figsize=(16,3))     # figure size

plt.subplot(1,2,1)             # subplot
plt.plot(x, y1, marker='o')    # plot in the same image
plt.plot(x, y2, marker='x')
plt.ylabel('y label')          # labels
plt.xlabel('x label')

plt.subplot(1,2,2)             
plt.plot(x, y1, marker='o')
plt.xlim([-5,15])              # axis range
plt.xticks(np.arange(-5,15,2.5))   # axis tick
print('Say something:')
```

## numpy datetime64
numpy datetime64 has units:
- ms, s, m, h, d, ...
- if not specified, the unit is inferred from the string.
- must be convert to int64, not int32.
- when converted to numbers, the unit is in effect.

```python
# datetime64
t1 = np.datetime64('1970-01-01T00:00:01', '[ms]')
t2 = np.datetime64('1970-01-01T00:00:01.000')     # same as t1

print('Unit is in effect when converted: {}ms'.format(t1.astype('int64')))
t2.astype('int64')    # convert to absolute value
```

```python
# plot and format datetime64
import matplotlib.dates as md

t = np.arange('2005-02', '2005-03', dtype='datetime64[D]')
y = np.random.random((28,))

plt.figure(figsize=(16,3))     # figure size
plt.plot(t, y, marker='o')
plt.xticks(rotation=75)       # datetime always need to rotate
ax=plt.gca()
xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')  # date time formatting
plt.xticks(t)                 # set all ticks visible
ax.xaxis.set_major_formatter(xfmt)
```
